```python
df1 = Faker()

def create_names(n):
    name = []
    for _ in range(0, n):
        name.append(df1.name())
    return name

name_list = create_names(362806)
df_name = pd.DataFrame(name_list, columns=['Name'])
```

```python
df_name.head()
```

```python
def weighted_cartesian_choice(*iterables):
    # An arbitrary number of tuple or lists,each consisting of population and weights.weighted_cartesian_choice returns a list
    #with a chocie from each population

    res = []
    for population, weights in iterables:
     #normalize weight:
        weights = np.array(weights) / sum(weights)
        lst = np.random.choice(population, p=weights)
        res.append(lst)
    return res
race = (['White','Hispanic', 'Black','Asian'],
        [.32,.28,.22,.18])

list=[]
for i in range(362806):
    res = weighted_cartesian_choice(race)
    list.append(res)
```

```python

df_race=pd.DataFrame(list, columns = ['Race'])
df_race.head()
```

# Split the data for training and testing

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.20)
```

##Scaling

```python
# import StandardScalar & create instance
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

# Model Building

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
model=clf.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
#find cnclusion matrix & to find accuracy,precision,recall,f1
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
print('Accuracy = ',accuracy_score(y_test,y_pred))
precision = precision_score(y_test, y_pred, average='weighted')
print('Precision = ', precision)
recall = recall_score(y_test, y_pred, average='weighted')
print('Recall score:', recall)
f1 = f1_score(y_test, y_pred, average='weighted')
print('F1 score:', f1)
```

```python
y_pred_rf=rf_clf.predict(x_test)
print('Accuracy = ',accuracy_score(y_test,y_pred_rf))
```

## XG Boost

```python
import xgboost as xgb
# Create the LabelEncoder object
label_encoder = LabelEncoder()

# Apply label encoding to the target variable
y_train_encoded = label_encoder.fit_transform(y_train)

# Create the XGBoost classifier
xgboost = xgb.XGBClassifier(learning_rate=0.1, max_depth=3, n_estimators=100)

# Fit the model on the training data
xgboost.fit(x_train, y_train_encoded)

# Predict the labels for the test data
y_pred_encoded = xgboost.predict(x_test)

# Inverse transform the predicted labels
y_pred = label_encoder.inverse_transform(y_pred_encoded)

# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

### Hyperparameter tuning

```python
from catboost import CatBoostClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

# Create the CatBoost classifier
catboost = CatBoostClassifier()

# Define the hyperparameters and their respective search space for tuning
param_grid = {
    'iterations': [100,150,200],
    'learning_rate': [0.01, 0.1,0.2],
    'depth': [4, 6,8]
}

# Initialize GridSearchCV with the CatBoost classifier and the hyperparameter grid
grid_search = GridSearchCV(estimator=catboost, param_grid=param_grid, cv=5)

# Fit the model on the training data using Grid Search with Cross-Validation
grid_search.fit(x_train, y_train)

# Get the best hyperparameters found by Grid Search
best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)

# Initialize CatBoost with the best hyperparameters
best_catboost = CatBoostClassifier(iterations=best_params['iterations'],
                                   learning_rate=best_params['learning_rate'],
                                   depth=best_params['depth'])

# Fit the model on the training data using the best hyperparameters
best_catboost.fit(x_train, y_train)

# Predict the labels for the test data using the best model
y_pred = best_catboost.predict(x_test)

# Calculate the accuracy of the best model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy with best model:", accuracy)
```
