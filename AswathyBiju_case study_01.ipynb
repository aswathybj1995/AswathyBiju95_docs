```python
# Linear Regression-Taking independent variable(Year) to x and dependent variable(Price) to y
```

```python
y=data['Price']
x=data.drop(['Price'],axis=1)
y
```

```python
x
```

```python
#Split the data for training and testing
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
# Creating instance for the linear model
```

```python
from sklearn import linear_model
lr=linear_model.LinearRegression()
```

```python
# Creating model
```

```python
model=lr.fit(x_train,y_train)
```

```python
# Test the model
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
# import MSE, r^2 to Evaluate the models with MSE & r^2 , print the scores
```

```python
from sklearn.metrics import mean_squared_error,r2_score
print('MSE = ',mean_squared_error(y_test,y_pred))
print('r2 score = ',r2_score(y_test,y_pred))
```

```python
# Lasso Regression importing from sklearn linear_model
```

```python
from sklearn import linear_model
ls=linear_model.Lasso()
```

```python
model1=ls.fit(x_train,y_train)
```

```python
y_pred1=model1.predict(x_test)
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error,r2_score
print('MSE = ',mean_squared_error(y_test,y_pred1))
print("r2 score= ",r2_score(y_test,y_pred1))
```

```python
# Predict the price of a 2022 model second-hand Grand i10 using Linear
```

```python
pred_price=model.predict([[2022]])[0]
print("Predicted price of a 2022 model second-hand Grand i10:", pred_price)
```

```python
# Predict the price of a 2022 model second-hand Grand i10 using Lasso
```

```python
pred_price1=model1.predict([[2022]])[0]
print("Predicted price of a 2022 model second-hand Grand i10:", pred_price1)
```
